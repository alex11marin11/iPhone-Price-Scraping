```python
def get_info(soup):
    
    # We find the info of the phone
    phone = soup.find('h1',class_=['D-QcMBXkOs9OFLIHLo84w']) 
    phone = phone.get_text() # Get only the text
    phone = phone.strip() # Get only the number
    phone = phone.split() # Convert into a list of words
    
    model = phone[0] + ' ' + phone[1] # Phone model
    memory = phone[2] # Phone memory capacity
    color = phone[5] # Phone color
    phone = ' '.join([model, memory, color]) # Final phone name
    
    # Find the price for each state
    prices = soup.find_all('div',class_="_1KnTWVHWxp0RXYEjGmT44-") # We find prices for each 5 aspects for the phone
    prices_clean = [] # List to store the prices
    for price in prices:
        price = price.get_text() # Get only the text
        price = price.strip() # Get only the number
        try:
            price = int(price[:-5]) # Get only the integer part and convert into int
        except ValueError:
            price = None
        prices_clean.append(price) # Append to the list of prices
        
    return model, int(memory), color, prices_clean

def append_to_csv(file_name, list_of_elem):
    '''
    Appends lists into csv file as rows
    '''
    
    with open(file_name, 'a+', newline='') as write_obj:    # Open file in append mode
        csv_writer = writer(write_obj)         # Create a writer object from csv module
        csv_writer.writerow(list_of_elem)        # Add contents of list as last row in the csv file


def insert_into_db(model, memory, color, state, daytime, price):
                
    # Create Query string
    sql = "INSERT INTO phone_price (phone, memory, color, state, daytime, price) VALUES (%s,%s,%s,%s,%s,%s)" # MySQL command
    val = (model, memory, color, state, daytime, price) # Values to put in the MySQL command

    try:
        mycursor.execute(sql, val) # Execute the query
        mydb.commit() # Commit to the query
        
    except:
        mydb.rollback() # Rollback in case there is any error

        
def check_phone(URL):
    
    # General info
    headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.198 Safari/537.36'}
    page = requests.get(URL, headers=headers) # Page where we will search the info
    soup = BeautifulSoup(page.content, 'html.parser') # We convert into a BeautifulSoup object
    csv_file = 'backup_DB.csv' # Csv file where we will save the info
    
    model, memory, color, prices = get_info(soup) # Get the info and price of the phone for all states
    daytime = datetime.now().strftime('%Y/%m/%d %H:%M:%S') # Get time in correct format
    
    if color == '(Product)Red': # We adapt the name of the color to our standard
        color = 'Rojo'

    for (price, state) in zip(prices, states):

        # Insert info into DB
        insert_into_db(model, memory, color, state, daytime, price) 
        
        # Append info to the csv file
        str_price = 'NULL' if price==None else str(price) # If price is None, change to 'NULL', if not keep value
        elements = ["('"+model+"'", memory, "'"+color+"'", "'"+state+"'", "'"+daytime+"'", str_price+")",''] # Elements 
        append_to_csv(csv_file, elements) # Call function to append elements into a row of the csv
```
